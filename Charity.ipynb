```python
# Годовая выручка компании "Завтрак" за 2021 год
revenue_2021 = 140983

# Годовая выручка компании "Завтрак" за 2022 год
revenue_2022 = 145424

# Вычисляем процентный рост выручки компании
growth_rate = (revenue_2022 / revenue_2021) - 1

# Годовая выручка компании на протяжении следующих 15 лет
revenues = [revenue_2022 * (1 + growth_rate) ** i for i in range(15)]

# Объем отчислений в благотворительный фонд "Завтраки для каждого" на протяжении следующих 15 лет
charity_contributions = [revenue * 0.05 for revenue in revenues]

# Выводим график отчислений в благотворительный фонд "Завтраки для каждого" на протяжении следующих 15 лет
import matplotlib.pyplot as plt

plt.plot(charity_contributions)
plt.title("Отчисления в благотворительный фонд \"Завтраки для каждого\" на протяжении следующих 15 лет")
plt.xlabel("Год")
plt.ylabel("Объем отчислений, руб.")
plt.show()
```
