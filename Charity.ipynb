```python
#    ""  2021 
revenue_2021 = 140983

#    ""  2022 
revenue_2022 = 145424

#     
growth_rate = (revenue_2022 / revenue_2021) - 1

#       15 
revenues = [revenue_2022 * (1 + growth_rate) ** i for i in range(15)]

#      "  "    15 
charity_contributions = [revenue * 0.05 for revenue in revenues]

#       "  "    15 
import matplotlib.pyplot as plt

plt.plot(charity_contributions)
plt.title("    \"  \"    15 ")
plt.xlabel("")
plt.ylabel(" , .")
plt.show()
```
